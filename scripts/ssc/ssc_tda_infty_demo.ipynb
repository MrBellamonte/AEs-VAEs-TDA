Generate data

```python
import tadasets
import gudhi as gd
import numpy as np

data = tadasets.infty_sign(n=35, noise = 0.07)

alpha_complex = gd.AlphaComplex(points = data)
alpha_simplex_tree = alpha_complex.create_simplex_tree()

rips_complex = gd.RipsComplex(points = data,max_edge_length=2)
rips_simplex_tree = rips_complex.create_simplex_tree(max_dimension=2)
```

Alpha-Filtration: Plot persistence diagrams to identify a "good" scale to plot simplicial complex after

```python
import matplotlib.pyplot as plt

diag_alpha = alpha_simplex_tree.persistence()
gd.plot_persistence_diagram(diag_alpha)
plt.show()
gd.plot_persistence_barcode(diag_alpha)
plt.show()
```

Plot alpha complex @ "good" scale

```python
from utils.plots import plot_simplicial_complex_2D
plot_simplicial_complex_2D(alpha_simplex_tree.get_skeleton(2), data, 0.18)
```

VR-Filtration: Plot persistence diagrams to identify a "good" scale to plot simplicial complex after

```python
import matplotlib.pyplot as plt

diag_alpha = rips_simplex_tree.persistence()
gd.plot_persistence_diagram(diag_alpha)
plt.show()
gd.plot_persistence_barcode(diag_alpha)
plt.show()
```
